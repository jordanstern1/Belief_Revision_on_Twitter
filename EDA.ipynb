```python
import pandas as pd
import numpy as np
import pickle 
from tweet_parser.tweet import Tweet
```

```python
# Load in large batch of tweets
fname = 'data/03_28_2018_18_02.pkl'
with open(fname, 'rb') as f:
    tweet_batch = pickle.load(f)
```

```python
#Remove duplicate (quote?) tweets
no_duplicates = []
text = []
for t in tweet_batch:
    if t.all_text not in text:
        no_duplicates += [t]
    text += [t.all_text]
    
len(no_duplicates)
```

```python
# make a pandas dataframe
df = pd.DataFrame([{"date": x.created_at_datetime,
                    "text": x.all_text,
                    "user": x.screen_name,
                    "bio": x.bio,
                    "profile_location": x.profile_location,
                    "quote_or_rt_text": x.quote_or_rt_text,
                    "at_mentions": [u["screen_name"] for u in x.user_mentions],
                    "hashtags": x.hashtags,
                    "urls": x.most_unrolled_urls,
                    "geo": x.geo_coordinates,
                    "type": x.tweet_type,
                    "id": x.id} for x in no_duplicates]).set_index("date")
```

```python
# look for specific issues
len([t for t in df.text if "gun" in t.lower() or "shoot" in t.lower() or "parkland" in t.lower() or "ar15" in t.lower()])

[t for t in df.text if "gitlab" in t.lower()]


```

# Tweets to Throw Out?

```python
# Want to throw out disengenuous tweets (not really expressing a change of opinion)
#Does "lol" imply sarcasm?

[t for t in df.text if "lol" in t]
```

```python
#How many tweets contain very little substantive content? Looks like very few! (could toss these)

min_length = len('changed my opinion')
short_length = min_length + 10 


short_tweets = [t for t in df.text if len(t) <= short_length]
print('number of very short tweets = ', len(short_tweets),'\n')
print('short tweets:\n',short_tweets)


```
